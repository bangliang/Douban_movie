```python
import requests
from lxml import etree
import  time

for a in range(10):
    
    url = 'https://movie.douban.com/top250?start={}'.format(a*25)
    data = requests.get(url).text
   
    s = etree.HTML(data) 
    file = s.xpath('//*[@id="content"]/div/div[1]/ol/li/div/div[2]')
    
    i = 1 + a*25
    
    for div in file:
        film = div.xpath('./div[1]/a/span[1]/text()')
        pf = div.xpath('./div[2]/div/span[2]/text()')
        num = div.xpath('./div[2]/div/span[4]/text()')
        scrible = div.xpath('./div[2]/p[2]/span/text()')
        time.sleep(1)
        
        if len(scrible) > 0:           
            print('%d.%s  %s  %s'%(i, film[0], pf[0], num[0]))
            print('      %s'%scrible[0])
            print('')
        else:
            print('%d.%s  %s  %s'%(i, film[0], pf[0], num[0]))
            print('')
            
        i += 1
```
